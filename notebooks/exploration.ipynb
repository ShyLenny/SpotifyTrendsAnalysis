```python
import matplotlib.pyplot as plt
```

```python
df["popularity"].hist(bins=20)

plt.title("Distribution of Song Popularity")
plt.xlabel("Popularity")
plt.ylabel("Number of Songs")

plt.savefig("../images/popularity_histogram.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
df[["energy", "popularity"]].head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

plt.scatter(df["energy"], df["popularity"], alpha=0.3)

plt.title("Popularity vs Energy")
plt.xlabel("Energy")
plt.ylabel("Popularity")

plt.savefig("../images/energy_vs_popularity.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.imshow(df.corr(numeric_only=True), cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns,
           rotation=90)

plt.yticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns)

plt.title("Correlation Heatmap")

plt.savefig("../images/correlation_heatmap.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
df.groupby("track_genre")["popularity"].mean()
```

```python
df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False)
```

```python
genre_popularity = (
    df.groupby("track_genre")["popularity"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

genre_popularity.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Genres by Average Popularity")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45)

plt.savefig("../images/top10_genres.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```
